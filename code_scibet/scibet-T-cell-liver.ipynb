```r
# start from here
data_transpose_count <- readRDS("data_transpose_count.rds")
label <- readRDS("anno.rds")
```

```r
for (i in 1:length(label$majorCluster)){
    if (is.na(label$majorCluster[i])){
        label$majorCluster[i] <- "unknown"
        print(i)
    }
}
```

```r
data_transpose_count$label <- label$majorCluster
```

```r
etest_gene <- SelectGene_R(data_transpose_count, k = 50)
etest_gene
```

```r
Marker_heatmap(data_transpose_count, etest_gene)
```

```r
tibble(
    ID = 1:nrow(data_transpose_count),
    label = data_transpose_count$label
) %>%
    dplyr::sample_frac(0.7) %>%
    dplyr::pull(ID) -> ID

train_set <- data_transpose_count[ID,]      #construct reference set
test_set <- data_transpose_count[-ID,]      #construct query set
```

```r
prob <- Train(train_set, k = 2000)
model.save <- data.frame(t(prob))
predict <- Test(prob, test_set)
```

```r
Confusion_heatmap(test_set$label, predict)
```
